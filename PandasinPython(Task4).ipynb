```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Gov.csv')

states = data['States and Union Territories']
phases = data['No. Of Phases']

colors = plt.cm.tab20.colors
bar_colors = [colors[i % len(colors)] for i in range(len(states))]

plt.figure(figsize=(12, 6))
bars = plt.bar(states, phases, color=bar_colors)

plt.xlabel('States and Union Territories')
plt.ylabel('Number of Phases')
plt.title('Lok Sabha Election Phases by State/UT')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```
