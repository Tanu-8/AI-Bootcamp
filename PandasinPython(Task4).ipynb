```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Gov.csv')

plt.figure(figsize=(14, 8))
plt.bar(df['States and Union Territories'], df['No. Of Phases'], color='pink')

plt.xlabel('States and Union Territories')
plt.ylabel('No. Of Phases')
plt.title('Number of Phases in Different States and UTs')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```
